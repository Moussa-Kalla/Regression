```python
from drawdata import ScatterWidget
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import random
```

```python
widget = ScatterWidget()
widget
```

```python
widget.data_as_pandas
widget.data_as_pandas.to_csv(r'C:\Python\Regression\Data\df.csv', index=False)
df = pd.read_csv(r'C:\Python\Regression\Data\df.csv')
plt.plot(df['x'], df['y'], 'o')
df
```

```python

x = df['x'].values.reshape(-1, 1)  
y = df['y'].values


degree = 3


poly_features = PolynomialFeatures(degree=degree)
x_poly = poly_features.fit_transform(x)


model = LinearRegression()
model.fit(x_poly, y)


y_pred = model.predict(x_poly)



plt.figure(figsize=(12, 7))
plt.scatter(x, y, label='Real Data', color='blue')
plt.plot(x, y_pred, color='red', label='Polynomial regression  degree' + str(degree))
plt.grid(True)
plt.title('Polynomial regression')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

# Calcul de l'erreur quadratique moyenne
mse = np.mean((y - y_pred) ** 2)
print(f'Mean Squared Error: {mse}')
```
