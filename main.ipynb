```python
# Transformer les listes en numpy arrays et redimensionner x
x = df['x'].values.reshape(-1,1)
y = df['y'].values.reshape(-1,1)

# Choisissez le degrÃ© du polynÃ´me (par exemple 2)
degree = 7

# Transformer les donnÃ©es en fonctionnalitÃ©s polynomiales
poly_features = PolynomialFeatures(degree=degree)
x_poly = poly_features.fit_transform(x)

# Ajuster le modÃ¨le de rÃ©gression linÃ©aire
model = LinearRegression()
model.fit(x_poly, y)

# PrÃ©dictions (peut Ãªtre utilisÃ© pour tracer la courbe de rÃ©gression)
y_pred = model.predict(x_poly)

plt.figure(figsize=(12, 7))
plt.plot(y_pred,color='r',label='RÃ©gression polynomiale de degrÃ© ' + str(degree))
plt.scatter(x,y, label='DonnÃ©es rÃ©elles')
plt.grid(True)
#plt.plot(xii, f(xii),color='g')
plt.title('RÃ©gression Polynomiale')
plt.xlabel('Variable indÃ©pendante x')
plt.ylabel('Variable dÃ©pendante y')
plt.grid(True)
plt.legend()
plt.show()

mse = np.mean((y - y_pred) ** 2)  
mse
```
